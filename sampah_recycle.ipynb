Query 1: Cek Kepatuhan Pajak Karbon
Deskripsi
Tujuan: Mengecek apakah emisi perusahaan dari CSV melebihi batas pajak karbon (50 ton CO2).

Konsep: If-else (Chapter 5), Pandas untuk baca CSV.

Output: Status pajak (kena/bebas) untuk setiap perusahaan.

Kode

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Baca data
df = pd.read_csv("sampah_recycle.csv")

# Tambahkan kolom persentase daur ulang
df['persentase_recycle'] = (df['didaur_ulang_kg'] / df['total_sampah_kg']) * 100

# Urutkan data berdasarkan tahun untuk memastikan garis tidak acak
plt.figure(figsize=(10, 6))

for provinsi in df['provinsi'].unique():
    data_prov = df[df['provinsi'] == provinsi].sort_values(by="tahun")
    plt.plot(data_prov['tahun'], data_prov['persentase_recycle'], marker='o', label=provinsi)

plt.title("Persentase Sampah Didaur Ulang per Tahun per Provinsi")
plt.ylabel("Persentase (%)")
plt.xlabel("Tahun")
plt.xticks(sorted(df['tahun'].unique()))  # pastikan urutan tahun di sumbu x rapi
plt.legend(title="Provinsi")
plt.tight_layout()
plt.show()
```
